# Example of using the drift_benchmark data module


```python
from drift_benchmark import data
```

## Baseline examples


```python
data_result = data.load_scenario("baselines/no_drift_synthetic")
columns = ["feature_1", "feature_2"]
data_result.X_ref[columns].plot(kind="hist", subplots=True, layout=(1, 2), figsize=(14, 3), bins=20)
data_result.X_test[columns].plot(kind="hist", subplots=True, layout=(1, 2), figsize=(14, 3), bins=20)
data_result.metadata.model_dump()
```

```python
data_result = data.load_scenario("baselines/no_drift_uci")
columns = ["sepal width", "petal length", "petal width"]
data_result.X_ref[columns].plot(kind="hist", subplots=True, layout=(1, 3), figsize=(14, 3), bins=20)
data_result.X_test[columns].plot(kind="hist", subplots=True, layout=(1, 3), figsize=(14, 3), bins=20)
data_result.metadata.model_dump()
```

## Synthetic example


```python
data_result = data.load_scenario("synthetic/concept_drift_gradual")
columns = ["feature_1", "feature_2"]
data_result.X_ref[columns].plot(kind="hist", subplots=True, layout=(1, 3), figsize=(14, 2), bins=20, xlim=(-3, 3))
data_result.X_test[columns].plot(kind="hist", subplots=True, layout=(1, 3), figsize=(14, 2), bins=20, xlim=(-3, 3))
data_result.metadata.model_dump()
```
