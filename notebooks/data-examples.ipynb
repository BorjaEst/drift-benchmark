# Example of using the drift_benchmark data module


```python
from drift_benchmark import data
```

## Scenario example


```python
data_result = data.create_sklearn_drift_scenario("iris_species_drift")
columns = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]
data_result.X_ref[columns].plot(kind="hist", subplots=True, layout=(1, 4), figsize=(15, 3), bins=20, xlim=(0, 8))
data_result.X_test[columns].plot(kind="hist", subplots=True, layout=(1, 4), figsize=(15, 3), bins=20, xlim=(0, 8))
pass
```

```python
data_result = data.load_dataset("example.csv")
columns = ["temperature", "min_temp", "humidity", "cloud_cover"]
data_result.X_ref[columns].plot(kind="hist", subplots=True, layout=(1, 4), figsize=(15, 3), bins=20, xlim=(0, 100))
data_result.X_test[columns].plot(kind="hist", subplots=True, layout=(1, 4), figsize=(15, 3), bins=20, xlim=(0, 100))
pass
```

## Dataset example


```python
data_result = data.generate_synthetic_data("GAUSSIAN", n_samples=100, n_features=6, drift_pattern="LINEAR")
columns = ["feature_1", "feature_2", "feature_3", "feature_4"]
data_result.X_ref[columns].plot(kind="hist", subplots=True, layout=(1, 4), figsize=(15, 3), bins=20, xlim=(-4, 4))
data_result.X_test[columns].plot(kind="hist", subplots=True, layout=(1, 4), figsize=(15, 3), bins=20, xlim=(-4, 4))
pass
```
